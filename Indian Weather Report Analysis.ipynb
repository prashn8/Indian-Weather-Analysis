INDIAN WEATHER REPORT ANALYSIS

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Checking the values of the data
df = pd.read_csv('IndianWeatherReport.csv')
df.head()
```

```python
#Checking the shape of the data
df.shape
```

```python
#Checking the info and if there is any null value in the data
df.info()
pd.isnull(df).sum()
```

```python
#Filter the data by the location name of the city
filtered_df = df[df['location_name'] == 'Rewa']
filtered_df
```

```python
#Calculating the average temperature in Celsius for Rewa
avg_temp_rewa = df['temperature_celsius'].mean()

# Calculating the maximum temperature in Celsius for Rewa
max_temp_rewa = df['temperature_celsius'].max()

# Calculating the minimum temperature in Celsius for Rewa
min_temp_rewa = df['temperature_celsius'].min()

print(f"The average temperature in Celsius for Rewa is {avg_temp_rewa:.2f}Â°C")
print(f"The maximum temperature in Celsius for Rewa is {max_temp_rewa:.2f}Â°C")
print(f"The minimum temperature in Celsius for Rewa is {min_temp_rewa:.2f}Â°C")
```

```python
# Group by 'region' and calculate mean temperature
region_mean_temp = df.groupby('region')['temperature_celsius'].mean()
region_mean_temp
```

```python
# Group by 'location_name' and calculate mean air quality
location_mean_air_quality = df.groupby('location_name')[['air_quality_PM2.5', 'air_quality_PM10']].mean()
location_mean_air_quality
```

```python
# Get the top 10 locations with the highest average air quality PM2.5
top_10_worst_air_quality = df.groupby('location_name')['air_quality_PM2.5'].mean().nlargest(10)

# Print the list of top 10 locations with the worst air quality
print("Top 10 Locations with Worst Air Quality (PM2.5):")
top_10_worst_air_quality
```

```python
# Top five hottest locations and temperatures
top_hottest = df.nlargest(5, 'temperature_celsius')[['location_name', 'temperature_celsius']]
print("Top five hottest locations and temperatures:")
top_hottest

```

```python

# Top five coldest locations and temperatures
top_coldest = df.nsmallest(5, 'temperature_celsius')[['location_name', 'temperature_celsius']]
print("\nTop five coldest locations and temperatures:")
top_coldest
```

```python
# Line plot of temperature over time for 'Rewa'
rewa_df = df[df['location_name'] == 'Rewa']
rewa_df['temperature_celsius'].plot(kind='line', title='Temperature Over Time for Rewa')
plt.show()
```

```python
# Box plot of air quality across different locations
sns.boxplot(x='region', y='air_quality_PM2.5', data=df)
plt.title('Air Quality PM2.5 Across Regions')
plt.show()
```

```python
sns.boxplot(x='region', y='temperature_celsius', data=df)
plt.title('Box Plot of Temperature by region')
plt.xlabel('Location')
plt.ylabel('Temperature (Celsius)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Violin plot of air quality PM2.5 across different locations
plt.figure(figsize=(20, 5))
sns.violinplot(x='region', y='air_quality_PM2.5', data=df)
plt.title('Air Quality PM2.5 Across Locations')
plt.xlabel('Location')
plt.ylabel('PM2.5')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin plot of air quality PM2.5 across different locations
plt.figure(figsize=(20, 5))
sns.violinplot(x='region', y='temperature_celsius', data=df)
plt.title('Temprature In Celsius Across Locations')
plt.xlabel('Location')
plt.ylabel('Temprature')
plt.xticks(rotation=45)
plt.show()
```

```python
# Get top 10 hottest and coldest locations with temperatures
top_10_hottest = df.nlargest(10, 'temperature_celsius')[['location_name', 'temperature_celsius']]
top_10_coldest = df.nsmallest(10, 'temperature_celsius')[['location_name', 'temperature_celsius']]

# Create a pie chart for the top 10 hottest locations and temperatures
plt.figure(figsize=(10, 5))
plt.pie(top_10_hottest['temperature_celsius'], labels=top_10_hottest['location_name'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Top 10 Hottest Locations and Temperatures')
plt.show()
```
